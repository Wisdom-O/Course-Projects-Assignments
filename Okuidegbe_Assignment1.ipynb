##### (2C)
Here, the data is bootsrapped in order to compare the mean of salary (Q10) for the two groups using a sample size equivalent to the number of data points in each group, and with sampling instances of 1000 .
After whch,the two bootstrapped distributions (for males and females) for the two groups is plotted  as well as the distribution of the difference in means.

```python
# selecting the data for the two groups from the dataframe as a pandas series object
female_salary=dQ2.loc[dQ2['Q2']=='Female']['Q10']
male_salary=dQ2.loc[dQ2['Q2']=='Male']['Q10']
```

```python
# converting the selected data from a pandas series to a numpy array for easy manipulation and calculation using numpy functions
female_salary_array=female_salary.to_numpy()
male_salary_array=male_salary.to_numpy()
```

```python
len(female_salary_array)
```

```python
type(male_salary_array)
```

```python
# bootstrapping the female salary data
female_salary_bootstrap=[]
female_n=len(female_salary_array)
for i in range(1000):
    female_salary_avg = np.mean(female_salary_array[np.random.randint(female_n,size=female_n)])
    female_salary_bootstrap.append(female_salary_avg)
```

```python
# bootstrapping the male salary data
male_salary_bootstrap=[]
male_n=len(male_salary_array)
for i in range(1000):
    male_salary_avg = np.mean(male_salary_array[np.random.randint(male_n,size=male_n)])
    male_salary_bootstrap.append(male_salary_avg)
```

```python
# plotting the distribution of the bootstrapped data. it can be seen that the distribution of the means of the data is normally
# distributed according to central limit theorem
plt.figure(figsize=(12,6))
plt.hist(female_salary_bootstrap,bins=20,label='female',color='blue')
plt.hist(male_salary_bootstrap,bins=20, label='male',color='purple')
plt.xlabel('Yearly Compensation/Salary',size=12)
plt.ylabel('Frequency of Occurrence',size=12)
plt.title('Distribution of the Bootstrapped data of the Yearly Salaries for Male and Female Survey Participants',size=14)
plt.legend()
```

```python
# Calculating the differences in mean of the two groups
female_salary_bootmean=np.mean(female_salary_bootstrap)
male_salary_bootmean=np.mean(male_salary_bootstrap)
diff_mean = mu = male_salary_bootmean - female_salary_bootmean

female_bootvar = np.var(female_salary_bootstrap)
male_bootvar = np.var(male_salary_bootstrap)
diff_var = (female_bootvar/female_n) + (male_bootvar/male_n)
stdev = np.sqrt(diff_var)
```

```python
# plotting the distribution for the difference in mean of the Salaries of the two groups
x=np.linspace(mu-(3*stdev),mu + (3 * stdev), 100)
plt.plot(x,stat.norm.pdf(x,mu,stdev))
plt.xlabel('Difference in means')
plt.title('Distribution of the Difference in means of the Salaries of the Two Groups')
```

The above curve shows the difference in mean of the distribtuion between the yearly salaries of male and female survey participants with a standard deviation of 50

#####  (2D) Hypothesis Testing
Since the bootstrapped distribution of the Salaries of the two groups is normally distributed, a 2-sample t-test can be performed on to check if the mean yearly salaries for each group is the same or different. The Null hypothesis for this test is that the means of the two groups are equal and the alternate hypothesis for this test is that the sampling distribution for the two groups are different. The p-value threshold to either reject or fail to reject the null hypothesis is set at 0.05. If the p-value of the test is greater than this threshold then the difference in mean of the two groups is not statistically significant, the Null hypothesis is True and the mean salaries of the bootstrapped data of these two groups is equal and vice versa. 

```python
tscore,pval = stat.ttest_ind(male_salary_bootstrap,female_salary_bootstrap)
if pval > 0.05:
    print('fail to reject null hypothesis')
else:
    print('reject null hypothesis')
print('test statistics is', tscore, '\nthe p-value is', pval)
```

After performing the 2-sample t-test for the bootstrapped distribution of the mean salary of each group, the test had a p-value of 0.0 which is less than 0.05, therefore, it can be inferred that the difference in mean salary of the two groups is statistically significant and we reject the Null hypothesis that the mean of the two groups are equal.

#### ANOVA

##### (3D)
In order to perform an ANOVA one way f test for the means of the bootstrapped distribution of the three groups, the assumptions made about the data for this test must be true. The assumptions made about the samples for the one way f test to be carried out are:    
Population distributions are normal,    
Samples have equal variances, and    
Independence among samples.  
From 3C, it was established that the bootstrapped means of the samples of the three groups are normally distributed. To find out if the samples have equal variances, a levene test would be performed on the three groups and if they do have the same variance, we move on to the next step.

The Levene test tests the null hypothesis that all input samples are from populations with equal variances with a significance level of 0.05. if the p-value of the test is less than 0.05, the null hypothesis is rejected in favour of the alternate hypothesis that all input samples are from populations with different variances. The levene test is performed below:

```python
statistics, p_val = stat.levene(bachelors_salary_bootstrap, masters_salary_bootstrap, doctoral_salary_bootstrap,center='mean')
if p_val > 0.05:
    print('Fail to Reject Null Hypthesis')
else:
    print('Reject Null Hypothesis')
print(statistics,p_val)
```

The p-value of the levene test is way less than 0.05, hence the null hypothesis is rejected in favour of the alternate hypothesis

Since the Levene tests performed revealed that all input samples are not from populations with equal variances, an ANOVA one way f test would not be performed as the result obtained will contain errors
